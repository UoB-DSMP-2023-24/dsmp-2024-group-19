```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from fast_tools import get_all_tapes
from tools import get_Tapes

# Load data
raw_tapes = get_Tapes(125,0)

tapes_data_1min = [t["Price"].resample("1min").last().ffill().to_numpy() for t in raw_tapes] # resample down to 1 min

# create numpy array    
data_array_1min = np.array(tapes_data_1min)



# 0 is time
# 1 is price
# 2 is volume
```

```python
# import hurst exponent function
from hurst import compute_Hc

# Create a list to store the results
hurst_results = []

# Loop through each tape
for i in range(len(data_array_1min)):
    # Compute the hurst exponent
    H, c, data = compute_Hc(data_array_1min[i], kind='price', simplified=True)
    # Append the results to the list
    hurst_results.append(H)
```

```python
# histogram of hurst exponents
plt.hist(hurst_results)
```
