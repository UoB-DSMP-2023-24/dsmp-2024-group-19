# import tapes and 15s features

```python
import pandas as pd
import numpy as np

# get lob data
raw_lob_data = pd.read_csv("final_eval_15.csv", index_col=0)
raw_lob_data = raw_lob_data.fillna(method = "ffill")
lob_data = []
for day in raw_lob_data["DAY"].unique():
    df = raw_lob_data[raw_lob_data["DAY"] == day]
    df.index = df["ENDT"]
    df = df.drop(columns = ["ENDT"])
    lob_data.append(df)

print(list(lob_data[0]))
lob_data = lob_data
lob_data[0]
```

# build trading environment with simple agent

```python
class Mean_Rev():
    def __init__(self, training_data, test_data):
        median = training_data["AP"].mean()

        self.indicators = pd.DataFrame(index = test_data.index)
        self.indicators["IND"] = 0
        
        for index in test_data.index:
            if median < test_data.loc[index,"close_hibid"]: # if the highest bid is > median it is a high price to sell
                self.indicators.loc[index,"IND"] = -1
            elif median > test_data.loc[index,"close_loask"]: # if the lowest ask is < median it is a low price to buy
                self.indicators.loc[index,"IND"] = 1
            else:
                pass

    def action(self, time_i):
        return self.indicators.loc[time_i]["IND"]

def evaluate_day(training_data, test_data, agent, starting_cash):
    trader = agent(training_data, test_data)

    agent_value = []
    agent_cash = []
    num_trades = 0

    cash = starting_cash
    assets = 0
    for i, row in test_data.iterrows():
        action = trader.action(i)

        if action == 1:
            if cash >= row["close_loask"]:
                cash -= row["close_loask"]
                assets += 1
                num_trades += 1

        elif action == -1:
            if assets > 0:
                assets -= 1
                cash += row["close_hibid"]
                num_trades += 1
        else:
            pass

        agent_value.append(cash + assets * row["AP"])
        agent_cash.append(cash)

    return agent_value, agent_cash, num_trades


def trading_env(level2_data: list[pd.DataFrame], agent):
    N = len(level2_data)
    starting_cash = 2_500

    buy_and_hold_profits = []
    agent_profits = []
    agent_sharpe = []
    buy_and_hold_sharpe = []
    agent_daily_values = []
    agent_num_trades = []
    agent_time_in_market = []

    for day_i in range(1,N):
        print(f"Testing day {day_i}", end = "\r")
        prev_day = level2_data[day_i - 1]
        curr_day = level2_data[day_i]

        agent_value, agent_cash, num_trades = evaluate_day(prev_day, curr_day, agent, starting_cash)

        agent_num_trades.append(num_trades)
        agent_time_in_market.append(1 - (np.mean(np.array(agent_cash) / np.array(agent_value))))

        agent_return = (agent_value[-1] - starting_cash) / starting_cash
        agent_profits.append(agent_return)
        day_return = (curr_day.iloc[-1]["AP"] / curr_day.iloc[0]["AP"]) - 1
        buy_and_hold_profits.append(day_return)
        agent_daily_values.append(agent_value)


        RF_rate = 0
        if agent_return == 0:
            agent_sharpe.append(0)
        else:
            agent_value = np.array(agent_value)
            ex_returns = (np.diff(agent_value) / agent_value[:-1]) - RF_rate
            agent_sharpe.append(np.mean(ex_returns) / np.std(ex_returns))

        ex_returns = curr_day["AP"].pct_change() - RF_rate
        buy_and_hold_sharpe.append(np.mean(ex_returns) / np.std(ex_returns))

    return [agent_profits, agent_sharpe, agent_daily_values, agent_num_trades, agent_time_in_market], [buy_and_hold_profits, buy_and_hold_sharpe]

results = trading_env(lob_data, Mean_Rev)
```

# Visualisation Function

```python
import matplotlib.pyplot as plt

def display_performance(results):

    # Assuming results is a list containing trader_performance, buy_and_hold_performance, sharpe_trader, and sharpe_buy_and_hold
    trader_performance = np.array(results[0][0])
    buy_and_hold_performance = np.array(results[1][0])
    sharpe_trader = results[0][1]
    sharpe_buy_and_hold = results[1][1]

    # Calculate cumulative returns
    trader_cumulative_returns = np.cumsum(trader_performance)
    buy_and_hold_cumulative_returns = np.cumsum(buy_and_hold_performance)

    # Create time axis
    x = np.arange(len(trader_performance))

    # Create cumulative Sharpe ratios
    cumulative_sharpe_trader = np.cumsum(sharpe_trader)
    cumulative_sharpe_buy_and_hold = np.cumsum(sharpe_buy_and_hold)

    # Create figure and axes
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # Plot cumulative returns on the first subplot
    axs[0].plot(x, trader_cumulative_returns, label='Trader', color='blue')
    axs[0].plot(x, buy_and_hold_cumulative_returns, label='Buy and Hold', color='red')
    axs[0].set_ylabel('Cumulative Returns')
    axs[0].set_xlabel('Days')
    axs[0].set_title('Cumulative Returns Over Time')
    axs[0].legend()

    # Plot cumulative Sharpe ratios on the second subplot
    axs[1].plot(x, cumulative_sharpe_trader, label='Trader', color='darkblue', linestyle='--')
    axs[1].plot(x, cumulative_sharpe_buy_and_hold, label='Buy and Hold', color='darkred', linestyle='--')
    axs[1].set_ylabel('Cumulative Sharpe Ratio')
    axs[1].set_xlabel('Days')
    axs[1].set_title('Cumulative Sharpe Ratio Over Time')
    axs[1].legend()

    # Adjust layout
    plt.tight_layout()

    # Show plot
    plt.show()

    trader_num_trades = results[0][3]
    trader_time_in_market = results[0][4]

    plt.figure(figsize=(14,3))
    plt.bar(x, trader_num_trades)
    plt.title('Number of Trades Over Time')
    plt.xlabel('Days')
    plt.ylabel('Number of Trades')
    plt.show()

    plt.figure(figsize=(14,3))
    plt.bar(x, trader_time_in_market)
    plt.title('Time Spent in Market Over Time')
    plt.xlabel('Days')
    plt.ylabel('Time in Market (days)')
    plt.show()


display_performance(results)
```

# build arima and lstm agents

```python
# importing libraries
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from sklearn import metrics
from sklearn.preprocessing import StandardScaler 
from keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split

class LSTM_agent():
    def __init__(self, training_data, test_data):
        self.indicators = pd.DataFrame(index = test_data.index)
        self.indicators["IND"] = 0

        scaler = StandardScaler()

        scaled_train = scaler.fit_transform(training_data["AP"].values.reshape(-1,1))
        scaled_test = scaler.transform(test_data["AP"].values.reshape(-1,1))

        window = 20
        X_train = []
        y_train = []
        for i in range(window, len(scaled_train)):
            X_train.append(scaled_train[i-window:i, 0])
            y_train.append(scaled_train[i, 0])

        X_test = []
        y_test = []
        for i in range(window, len(scaled_test)):
            X_test.append(scaled_test[i-window:i, 0])
            y_test.append(scaled_test[i, 0])

        # The data is converted to Numpy array
        X_train, y_train = np.array(X_train), np.array(y_train)
        X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1],1))
        y_train = np.reshape(y_train, (y_train.shape[0],1))

        # The data is converted to numpy array
        X_test, y_test = np.array(X_test), np.array(y_test)
        X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1],1))
        y_test = np.reshape(y_test, (y_test.shape[0],1))

        regressorLSTM = Sequential()
        regressorLSTM.add(LSTM(16, 
                            return_sequences = True, 
                            input_shape = (X_train.shape[1],1)))
        regressorLSTM.add(LSTM(8, 
                            return_sequences = False))
        #regressorLSTM.add(Dense(2))
        regressorLSTM.add(Dense(1))
        
        regressorLSTM.compile(optimizer = 'adam',
                            loss = 'mean_squared_error',
                            metrics = ["mean_absolute_error"])
        
        #Fitting the model
        early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

        X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

        # Fitting the model with early stopping
        regressorLSTM.fit(X_train, 
                        y_train, 
                        batch_size = 64, 
                        epochs = 100, 
                        validation_data=(X_val, y_val),
                        callbacks=[early_stopping])

        regressorLSTM.save(f"MODELS/LSTM_{training_data['DAY'].mean()}.keras")

        y_LSTM = regressorLSTM.predict(X_test)
        y_LSTM_O = scaler.inverse_transform(y_LSTM)
        pred2 = np.array(y_LSTM_O)
        padded_pred2 = np.pad(np.ravel(pred2), (window, 0), mode='constant')
        assert len(padded_pred2) == len(test_data)
        
        self.indicators["LSTM_pred"] = padded_pred2

        for index in test_data.index:
            if self.indicators.loc[index, "LSTM_pred"] < test_data.loc[index,"close_hibid"]: # if the highest bid is > median it is a high price to sell
                self.indicators.loc[index,"IND"] = -1
            elif self.indicators.loc[index, "LSTM_pred"] > test_data.loc[index,"close_loask"]: # if the lowest ask is < median it is a low price to buy
                self.indicators.loc[index,"IND"] = 1
            else:
                pass

    def action(self, index):
        return self.indicators.loc[index,"IND"]
    
display_performance(trading_env(lob_data, LSTM_agent))
```
