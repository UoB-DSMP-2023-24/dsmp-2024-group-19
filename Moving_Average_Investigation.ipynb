```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tools import *
```

```python
tape1 = pd.read_csv("resampled_df.csv")
```

```python
print(tape1.head())
```

```python
plt.figure(figsize=(20,10))
plt.plot(tape1['Price'], color='blue', label='Price')
plt.show()
```

```python
def moving_average(data, window):
    return np.convolve(data, np.ones(window), 'valid') / window

tape1_SMAs_20 = moving_average(tape1['Price'], 20)
tape1_SMAs_50 = moving_average(tape1['Price'], 50)
tape1_SMAs_100 = moving_average(tape1['Price'], 100)
```

```python
plt.figure(figsize=(20,10))
plt.plot(tape1['Price'], label='Price')
plt.plot(tape1_SMAs_20, label='20-day SMA', color='orange')
plt.plot(tape1_SMAs_50, label='50-day SMA', color='red')
plt.plot(tape1_SMAs_100, label='100-day SMA', color='green')
```
